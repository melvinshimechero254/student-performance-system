# TUM Student Performance — Data Simulation
Generates realistic student data matching TUM semester structure:
- CATs out of 30 marks (2–4 per semester)
- Final exam out of 70 marks
- Attendance as percentage
- Labels: Pass / At Risk / Fail


```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 600  # simulated students

# Attendance: realistic distribution, most students 60-95%
attendance = np.clip(np.random.normal(75, 15, n), 0, 100).astype(int)

# CAT1 out of 30 — month 2. Correlated with attendance.
cat1 = np.clip(15 + (attendance - 75) * 0.25 + np.random.normal(0, 5, n), 0, 30).round(1)

# CAT2 out of 30 — month 3. Slight improvement possible.
cat2 = np.clip(cat1 + np.random.normal(1, 4, n), 0, 30).round(1)

# CAT average (best-of scoring)
cat_avg = ((cat1 + cat2) / 2).round(2)

# Assignment score as percentage
assignment_pct = np.clip(70 + (attendance - 75) * 0.4 + np.random.normal(0, 12, n), 0, 100).round(1)

# Assignments submitted percentage — engagement indicator
assignments_submitted = np.clip(attendance * 0.85 + np.random.normal(5, 10, n), 0, 100).round(1)

# Final exam out of 70 — simulated outcome (NOT used as model feature)
final_exam = np.clip(
    (cat_avg / 30) * 35 + (attendance / 100) * 20 +
    (assignment_pct / 100) * 10 + np.random.normal(0, 6, n),
    0, 70
).round(1)

# Labels based on simulated final exam outcome
labels = []
for i in range(n):
    f, a, c = final_exam[i], attendance[i], cat_avg[i]
    if a < 65 or f < 28:
        labels.append('Fail')
    elif a < 70 or f < 35 or c < 12:
        labels.append('At Risk')
    else:
        labels.append('Pass')

df = pd.DataFrame({
    'Attendance': attendance,
    'CAT1_Score': cat1,
    'CAT2_Score': cat2,
    'CAT_Average': cat_avg,
    'Assignment_Score': assignment_pct,
    'Assignments_Submitted': assignments_submitted,
    'Final_Exam': final_exam,
    'Performance': labels
})

print(df['Performance'].value_counts())
df.to_csv('../data/student_performance.csv', index=False)
print('Saved to data/student_performance.csv')
```
